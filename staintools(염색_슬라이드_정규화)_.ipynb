```python
import numpy as np
import matplotlib.pyplot as plt
import staintools
import spams
```

```python
target = staintools.read_image("/content/0006ff2aa7cd.tif")
to_transform = staintools.read_image("/content/00a768c3dfa5.tif")
plt.subplot(1,2,1)
plt.title('target')
plt.imshow(target)
plt.subplot(1,2,2)
plt.title('to_transform')
plt.imshow(to_transform)
```

```python
target = staintools.LuminosityStandardizer.standardize(target)
to_transform = staintools.LuminosityStandardizer.standardize(to_transform)
plt.subplot(1,2,1)
plt.title('target')
plt.imshow(target)
plt.subplot(1,2,2)
plt.title('to_transform')
plt.imshow(to_transform)
```

```python
normalizer = staintools.StainNormalizer(method='vahadane')
normalizer.fit(target)
transformed = normalizer.transform(to_transform)
plt.subplot(1,2,1)
plt.title('to_transform')
plt.imshow(to_transform)
plt.subplot(1,2,2)
plt.title('transformed')
plt.imshow(transformed)
```
